```python
import pandas as pd
import numpy as np

print('Pandas:', pd.__version__)
print('Numpy:',np.__version__)

df = pd.read_csv('/Users/hibbert/Desktop/homework/MLP/hw1/train-data.csv',index_col = 0) # read in the csv file

df.head()
```

```python
df['Name'] = df.Name.transform(lambda t: ''.join([x, ' '][x in '.,'] for x in t).split()[0])
df.describe()
df.head()
```

```python
df['Pow'] = df['Power'].map(lambda x: x.rstrip(' bhp'), na_action='ignore')
df['Pow'] = df['Pow'].map(lambda x: x.rstrip('null'), na_action='ignore')
df['Pow'] = pd.to_numeric(df['Pow'])
df['Pow'] = df['Pow'].astype(float, errors = 'ignore')
df.describe()
df.head()
```

```python
df['Eng'] = df['Engine'].map(lambda x: x.rstrip(' CC'), na_action='ignore')
df['Eng'] = pd.to_numeric(df['Eng'])
df.describe()
df.head()
```

```python
#df.Name = df.Name.map( {'Maruti':0 , 'Hyundai':1, 'Honda':2, 'Audi':3, } )
#TODO map name to index
```

```python
df.Fuel_Type = df.Fuel_Type.map( {'CNG':0 , 'Diesel':1, 'Petrol':2, 'LPG':3} )
df.Fuel_Type = pd.to_numeric(df.Fuel_Type)
df.head()
```

```python
df.Transmission = df.Transmission.map( {'Manual':0 , 'Automatic':1} )
df.Transmission = pd.to_numeric(df.Transmission)
df.head()
```

```python
df.Owner_Type = df.Owner_Type.map( {'First':0 , 'Second':1, 'Third':2, 'Fourth & Above':3} )
df.Owner_Type = pd.to_numeric(df.Owner_Type)
df.head()
```

```python
#df['Mage'] = df['Mileage'].map(lambda x: x.rstrip(' kmpl'), na_action='ignore')
#df['Mage'] = df['Mage'].map(lambda x: 1.6747*pd.to_numeric(x.rstrip(' km/kg')), na_action='ignore')
#df['Mage'] = df['Mage'].astype(float, errors = 'ignore')
df = df.dropna(subset=['Mileage'])
df['Mage'] = df['Mileage'].map(lambda x: x.rstrip(' kmpl'), na_action='ignore')
mask = df['Mage'].str.endswith(' km/kg')
df['Mage'].loc[mask] = pd.to_numeric(df['Mage'].loc[mask].str.replace(' km/kg',''))*1.6747
df['Mage'] = pd.to_numeric(df['Mage'])
df.describe()
df.head()
```

Impute

```python
from sklearn.impute import KNNImputer
import copy

# get object for imputation
knn_obj = KNNImputer(n_neighbors=5)

# create a numpy matrix from pandas numeric values to impute
temp = df[['Year','Kilometers_Driven','Price','Pow','Eng','Mage']].to_numpy()

# use sklearn imputation object
knn_obj.fit(temp)
temp_imputed = knn_obj.transform(temp)
#    could have also done:
# temp_imputed = knn_obj.fit_transform(temp)

# this is VERY IMPORTANT, make a deep copy, not just a reference to the object
df_imputed = copy.deepcopy(df) # not just an alias
df_imputed[['Year','Kilometers_Driven','Price','Pow','Eng','Mage']] = temp_imputed
df_imputed.info()
```

```python
import matplotlib.pyplot as plt
import warnings
import seaborn as sns

warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline 

# plot the correlation matrix using seaborn
sns.set(style="darkgrid") # one of the many styles to plot using

cmap = sns.diverging_palette(220, 10, as_cmap=True)

f, ax = plt.subplots(figsize=(9, 9))

sns.heatmap(df.corr(), cmap=cmap, annot=True)

f.tight_layout()
plt.title('Correlation Matrix Graph')
```

According to the correlation table, "Price", "Pow","Eng", and "time_spend_company" have positive correlation with each other, along with "Eng" and "Seats", each with positive correlation rate that is more than 0.1. It seem that engine, power and price, and engine with the number of seats are strongly related to each other. Since our lab focus on the price, we can see in the table that Price & Pow and Price & Eng have comparatively strong negative correlation with each other, that means the stronger the Engine and the power are , the higher the second price will be. Seats & Price has postive correlation with each other, that means the more seats the car have, the higher its price.
